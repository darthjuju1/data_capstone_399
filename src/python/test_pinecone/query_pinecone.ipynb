```python
import re
import pandas as pd

def parse_year_field(s):
    if pd.isna(s):
        return pd.NaT
    s = str(s).strip()
    # 4‑digit year
    if re.fullmatch(r"\d{4}", s):
        return pd.to_datetime(f"{s}-01-01")
    # year-month
    if re.fullmatch(r"\d{4}-\d{1,2}", s):
        # ensure zero‑padded month
        year, month = s.split("-")
        month = month.zfill(2)
        return pd.to_datetime(f"{year}-{month}-01")
    # everything else: let pandas try to infer
    return pd.to_datetime(s, infer_datetime_format=True, errors="coerce")

df['Year'] = df['Year'].apply(parse_year_field)

```

```python

records = (
    df[['_id', 'Lyrics', 'Genre', 'Year', 'Track Name', 'Artist Name']]
      .rename(columns={
         'Lyrics': 'chunk_text',      # this is what gets embedded
         'Genre' : 'genre',        # metadata
         'Year'  : 'year',            # metadata
         'Track Name' : 'track_name',      # metadata
         'Artist Name': 'artist_name'      # metadata
      })
    .to_dict(orient='records')
)

records
```

```python
for rec in records:
    try:
        test_index.upsert_records("test-namespace", [rec])
        print(f"Record {rec['_id']} upserted successfully.")
    except Exception as e:
        print(f"Error upserting record {rec['_id']}: {e}")
```

```python
# Search with a query text and rerank the results
# Supported only for indexes with integrated embedding
search_with_text = test_index.search(
    namespace="test-namespace", 
    query={
        "inputs": {"text": "Furry"}, 
        "top_k": 4
    },
    fields=["category", "chunk_text"],
    rerank={
        "model": "bge-reranker-v2-m3",
        "top_n": 2,
        "rank_fields": ["chunk_text"] # Specified field must also be included in 'fields'
    }
)

print(search_with_text)
```
